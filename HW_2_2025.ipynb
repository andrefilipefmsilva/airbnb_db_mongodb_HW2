**Decision**: Keep host related information nested inside the listing document. Points considered and justification:
- **Host related operations vs. listing related operations**: Usually users search by location, price, amenities, so the main logic revolves around listing. The interest for a specific host information comes as secondary from the interest in the listing, meaning that most of the time we will retrieve the listing and need complementary information on the host, rather than retrieve the host and then look at the listings he has. A separate host document would only be beneficial if querying all listings for a specific host is common, since we have more listings related operations, having the host informations inside the listing document ensures no need for additional queries across different collections.

- **Frequency of updating host fields**: Having separate host documents in another collection makes updating fields easier, because we only have to update one document, instead of updating all listing documents where that host is. However, most fields are unlikely to be updated frequently, host information often remains the same,  so keeping them inside the listing document prevents unnecessary complexity. 

- **Duplicated information**: Separate host document prevents duplication, however one listing only has one host and keeping the host information inside the listing does not cause the document to be very long, as such the duplication may be justified considering the previous points. On average each host has 15 listings so the duplication is not very extreme, but we recognize there are some cases where the duplication will be heavy (e.g. max value of 1198).

**Decision**: Separate transactions and listings into two collections, so that transaction-specific data can be managed independently. Points considered and justification:

- **Size of the document**: Currently we have on average 56 transactions nested inside the listing's documents and transaction data tends to grow over time as more transactions occur for each listing. This results in larger and slower to query documents, with a risk of hitting MongoDB’s document size limit. By moving transactions into a separate collection, we can keep the listing documents smaller and more manageable.

- **Listing-related operations vs. transaction-related operations**: The primary use case is to display property listings to customers, who generally do not need past transaction data. Including transaction data in the listings collection could slow down queries focused on listings. By having a separate transactions collection, listing-related queries can remain fast and efficient and transaction-related queries can be performed independently on the transactions collection.

```python
listings_collection = db["listingsAndReviews_new"]
reviews_collection = db["Reviews"]
listings_new_collection = db["Listings"]

def migrate_reviews():
    for listing in listings_collection.find():
        listing_id = listing["_id"]
        reviews = listing.get("reviews", [])

        if reviews:
            # Order reviews by date
            reviews_sorted = sorted(reviews, key=lambda x: x["date"], reverse=True)

            for review in reviews:
                review["listing_id"] = listing_id
                reviews_collection.update_one(
                    {"_id": review["_id"]},
                    {"$set": review},
                    upsert=True
                )

            # keep 3 most recent reviews
            Latest_reviews = reviews_sorted[:3]
        else:
            Latest_reviews = []

        listings_new_collection.insert_one({
            "_id": listing["_id"],
            **{k: v for k, v in listing.items() if k != "reviews"},
            "Latest_reviews": Latest_reviews
        })

    db.listingsAndReviews.drop()
    print("Migration completed and old collection dropped")

migrate_reviews()
```


**Patterns used**:

- **Bucket**: Bucket was already being used for the transactions related data, however smaller buckets help limit the size of the data that needs to be processed at once. We believe that it’s more common to analyze transactions by year because properties are usually rented for days/weeks, so we created buckets for each year rather than having a single bucket that contains data across several years. This approach enhances performance, as calculations typically only require the most recent year’s data. 

- **Computed**: Add computed value average_rating for quick retrieval. Since this is a very common value to be retrieved in this type of platform, it is justified that this value is precomputed and saved in the dataset, rather than having to compute it each time that we need to display the value. 

- **Subset**:
We created a Host field which will have a subset with every host related fields that were already present in the database. By structuring the data this way, we can easily access all host-related information, rather than having to search through multiple individual host fields. 

- **Extended reference**:
    - We applied extended reference when creating the Reviews collection, where each review contains the listing_id, and each listing contains an array of review IDs. This will be useful in cases where the user wants to check all reviews for a given listing, or at least more than the ones stored in the listing document. If we only stored listing_id in the reviews, we would need to query the reviews collection to gather all the reviews for a particular listing. However, having the array of review_ids in the listings collection allows to quickly retrieve all reviews for a listing without having to query the reviews collection. 
    - However given that transaction data is not commonly retrieved in most queries, because the main concern is displaying listing information to users, we believe it’s not justified to use extended references in the listing documents for this case.

### Question 6 - Identify areas by whether they are typically used for short breaks, like weekend mini breaks, or whether they are more suitable for long trips. This information support targeted advertising of different customer types. It is not expected to change much over time so we won’t look to update it, we just require current view. What is the average duration of stay (in nights) per type of property per city (you can use the maximum_nights to measure length of stays)? For each property type return the city with the highest and lowest average value.

```python
pipeline = [
    # Ensure maximum_nights exists and is a valid number
    {"$match": {"maximum_nights": {"$exists": True, "$ne": ""}
    }},


    # Group by city and property type, calculating the average maximum_nights
    {"$group": {
        "_id": {"city": "$address.market", "property_type": "$property_type"},
        "avg_maximum_nights": {"$avg": {"$toDouble": "$maximum_nights"}}
    }},

    # Reshape the output for easier processing
    {"$project": {
        "city": "$_id.city",
        "property_type": "$_id.property_type",
        "avg_maximum_nights": 1,
        "_id": 0
    }},

    # Group by property type to find the cities with the highest and lowest average duration
    {"$group": {
        "_id": "$property_type",
        "highest_city": {"$max": {"city": "$city", "avg_nights": "$avg_maximum_nights"}},
        "lowest_city": {"$min": {"city": "$city", "avg_nights": "$avg_maximum_nights"}}
    }}
]

# Execute query
result = list(db.listingsAndReviews.aggregate(pipeline))

# Print result
for item in result:
    pprint(item)
```

```python
collection = db["listingsAndReviews"]
suggested_amenities = db["suggested_amenities"]

# Extract unique amenities
unique_amenities = collection.aggregate([
    {"$unwind": "$amenities"},
    {"$group": {"_id": "$amenities"}}
])

# Store the unique amenities in a single document
amenities_list = sorted(set(doc["_id"] for doc in unique_amenities))
suggested_amenities.update_one(
    {"_id": "amenities_list"},
    {"$set": {"amenities": amenities_list}},
    upsert=True
)
```
