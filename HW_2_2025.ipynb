### Hosts:

```python
pipeline = list(
    db.listingsAndReviews.aggregate([
        {"$project": {"host": 1 }},
        {"$limit": 1}
    ]))

pprint(pipeline)
```

Check nº of hosts, max and average listings per host - to understand degree of duplication:

```python
# Total nº of hosts 

pipeline = list( 
    db.listingsAndReviews.aggregate([ 
        {"$group": {"_id": "$host.host_id"}}, 
        {"$count": "total_hosts"},
        {"$project": {"_id": False}} 
        ] ) ) 

pprint(pipeline)
```

```python
# Total number of listings per host - top 10

pipeline = list( 
    db.listingsAndReviews.aggregate([ 
        {"$group": {"_id": "$host.host_id", "nr_listings": {"$max": "$host.host_total_listings_count"}}}, 
        {"$sort": {"nr_listings": -1}}, 
        {"$limit": 10},
        {"$project": {"_id": False}} 
        ] ) ) 

pprint(pipeline)
```

```python
# Average number of listings per host

pipeline = list( 
    db.listingsAndReviews.aggregate([ 
        {"$group": {"_id": None, "avg_listings": {"$avg": "$host.host_total_listings_count"}}},
        {"$project": {"_id": False}} 
        ] ) ) 

pprint(pipeline)
```

**Decision**: Keep host related information nested inside the listing document. Points considered and justification:
- **Host related operations vs. listing related operations**: Usually users search by location, price, amenities, so the main logic revolves around listing. The interest for a specific host information comes as secondary from the interest in the listing, meaning that most of the time we will retrieve the listing and need complementary information on the host, rather than retrieve the host and then look at the listings he has. A separate host document would only be beneficial if querying all listings for a specific host is common, since we have more listings related operations, having the host informations inside the listing document ensures no need for additional queries across different collections.

- **Frequency of updating host fields**: Having separate host documents in another collection makes updating fields easier, because we only have to update one document, instead of updating all listing documents where that host is. However, most fields are unlikely to be updated frequently, host information often remains the same,  so keeping them inside the listing document prevents unnecessary complexity. 

- **Duplicated information**: Separate host document prevents duplication, however one listing only has one host and keeping the host information inside the listing does not cause the document to be very long, as such the duplication may be justified considering the previous points. On average each host has 15 listings so the duplication is not very extreme, but we recognize there are some cases where the duplication will be heavy (e.g. max value of 1198).

**Decision**: Separate reviews and listings in two collections, so we can have review specific documents. Points considered and justification: 
- **Size of the document**: The listings have on average 27 reviews, with some cases having hundreds of reviews nested in a single listing. The field “comments” often holds values that are long lines of text. Additionally, with time this problem only tends to get worse as new reviews are added. Given this scenario it would be beneficial to have a new collection for review documents.
- **Review related operations vs. listing related operations**: Usually a user will be interested in a specific listing and want to look at reviews related to that listing. This means that a separate reviews document will lead to additional queries across collections in some cases. However, it is also common to have review related queries (e.g. average review score per listing), and if reviews have a separate collection we will only need to load review related information, avoiding loading all the remaining information that is in the listing’s documents.
- **Keep some info in the listing's documents**: Aditionally, we will keep 3 reviews nested inside the listings' documents for efficient querying.
- **Implications**: We need to ensure consistency between collections. For example, when a new review is added we need to update some fields in the listings' documents, like: last_review and number_of_reviews.

### Question 2 - Once a month, we reward hosts with recognition. Select three superhosts with at least two listings that can accommodate more than four people.

```python
pipeline = list(
    db.listingsAndReviews.aggregate([
        {"$match": { 
            "host.host_is_superhost": True,
            "accommodates": { "$gt": 4 }
        }},
        {"$group": { 
            "_id": "$host.host_id", 
            "host_name": { "$first": "$host.host_name" }, 
            "host_listings_count": { "$sum": 1 }
        }},
        {"$match": { 
            "host_listings_count": { "$gte": 2 } 
        }},
        {"$sort": { "host_listings_count": -1 }}, 
        {"$limit": 3}
    ])
)

print("\nSelected Superhosts:\n")
for doc in pipeline:
    pprint(doc)
```

### Question 3 - The company considers investing into property to rent. Which bed type is most common in listings with a waterfront and a dishwasher in New York?

```python
pipeline = list(
    db.listingsAndReviews.aggregate([
        {"$match": { 
            "address.market": "New York",
            "amenities": {"$in": ["Dishwasher", "Waterfront"]},
        }},
        {"$group": {
            "_id": "$bed_type", 
            "count": {"$sum": 1}
        }},
        {"$sort": {"count": -1}},
        {"$limit": 1}
    ])
)

for doc in pipeline:
    pprint(doc)
```

### Question 4 - We're considering hiring someone to write reviews professionally. Who wrote the longest review in New York?

```python
pipeline = list(
    db.listingsAndReviews.aggregate([
        {"$match": { 
            "address.market": "New York", 
            "reviews": { "$exists": True } 
        }},
        {"$unwind": "$reviews"},
        {"$sort": { "reviews.comments": -1 }}, 
        {"$project": { 
            "reviews.reviewer_name": 1,
            "reviews.comments": 1
        }},
        {"$limit": 1}
    ])
)

print("\nPerson with the Longest Review in New York:\n")
for doc in pipeline:
    pprint(doc)
```

### Question 5 - To assess the security of different areas, what is the biggest and smallest (price-security deposit) difference per number of visitors at a property?

```python
pipeline = list(
    db.listingsAndReviews.aggregate([
        {"$match": {
            "price": {"$exists": True},
            "security_deposit": {"$exists": True}  
        }},

        {"$project": {
            "difference": {"$subtract": ["$price", "$security_deposit"]} 
        }},
        {"$group": {
            "_id": None, 
            "max_difference": {"$max": "$difference"}, 
            "min_difference": {"$min": "$difference"}
        }}
    ])
)

print("\nBiggest and Smallest Price-Security Deposit Differences:\n")
for doc in pipeline:
    pprint(doc)

```

```python
#mesma coisa mas a retornar os ID's 
pipeline = list(
    db.listingsAndReviews.aggregate([
        {"$match": {
            "price": {"$exists": True},  
            "security_deposit": {"$exists": True}  
        }},
        {"$project": {
            "_id": 1,  
            "price": {"$toDouble": "$price"},  
            "security_deposit": {"$toDouble": "$security_deposit"},  
            "difference": {"$subtract": [{"$toDouble": "$price"}, {"$toDouble": "$security_deposit"}]}  
        }},
        {"$sort": {"difference": 1}}, 
        {"$group": {
            "_id": None,  
            "max_difference": {"$last": "$difference"},  
            "max_id": {"$last": "$_id"},
            "min_difference": {"$first": "$difference"},  
            "min_id": {"$first": "$_id"} 
        }}
    ])
)

print("\nBiggest and Smallest Price-Security Deposit Differences:\n")
for doc in pipeline:
    pprint(doc)
```
